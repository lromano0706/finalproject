```python
import pandas as pd
```

```python
cities_df = pd.read_csv("Data/cal_cities_lat_long.csv")
cities_df = cities_df.rename(columns={"Name": "City"})
cities_df.head()
```

```python
crime_cities_df = pd.read_csv("Data/OC_crime.csv")
crime_cities_df.head()
```

```python
crime_cities_coords_df = cities_df.merge(crime_cities_df, how='inner', on='City').reset_index(drop = True)
```

```python
crime_cities_coords_df.head()
```

```python
crime_cities_coords_df = crime_cities_coords_df[['Department_Name', 'State', 'Ori', 'City', 'Latitude', 'Longitude' , 'Agg_Assult', 'Sex_Offences', 'Manslaughter_Neg','Murder_and_Nonneg_Man','Rape','Robbery','Simp_Assult','Humman_Trafficking_Commercial_SA','Humman_Trafficking_Invol_Ser', 'Offences_Total']]
crime_cities_coords_df.head()
```

```python
crime_cities_coords_df.to_csv(r'Data\crime_cities_coords.csv', index = False)
```
