```python
voice = pd.read_csv('../Resources/voice.csv')
voice.head()
```

## Data Pre-Processing

```python
#step 1: choose columsn of interest
#step 2: use get_dummies
#step 3: follow this now voice_recognition
X = voice.drop(["label","Q25"], axis=1)
y = voice["Q25"].values
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.keras.utils import to_categorical
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=1)
```

```python
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

# Create a Deep Learning Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Create model and add layers
model = Sequential()
model.add(Dense(units=100, activation='relu', input_dim=19))
model.add(Dense(units=100, activation='relu'))
# model.add(Dense(units=2, activation='softmax'))
model.add(Dense(units=1))
```

```python
# Compile and fit the model
model.compile(optimizer='adam',
              loss='mean_squared_error',
              metrics=['mse'])
```

```python
model.summary()
```

```python
model.fit(
    X_train_scaled,
    y_train,
    
    validation_split=.15,

    epochs=17,
    shuffle=True,
    verbose=2
)
```

## Quantify our Trained Model

```python
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

## Make Predictions

```python
prediction_labels = model.predict_classes(X_test_scaled)
print(prediction_labels)
#prediction_labels = label_encoder.inverse_transform(encoded_predictions)
```
